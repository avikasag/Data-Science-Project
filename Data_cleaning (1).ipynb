```python
import pandas as pd
df = pd.read_excel('Customer Call List.xlsx')
print(df.head(20))
```

```python
df = df.drop_duplicates()
print("\nDataFrame after removing duplicates:")
print(df)
```

```python
#dropping the table with name='Not_useful_column'
df=df.drop(columns='Not_Useful_Column')
```

```python
#dropping the ... from the last_name
df["Last_Name"]=df["Last_Name"].str.lstrip('...')
#dropping the / from the last_name
df["Last_Name"]=df["Last_Name"].str.lstrip('/')
#droping the _ from the Last_name using _
df["Last_Name"]=df["Last_Name"].str.strip('_')
```

```python
df
```

```python
# Replacing all non-numeric and non-alphabetic characters with empty string
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[^a-zA-Z0-9]', '', regex=True)
print(df)
```

```python
# Remove non-numeric rows
df = df[df['Phone_Number'].str.match(r'^\d+$', na=False)]
df
```

```python
#applying lambda on the phone (To convert numeric to string data type)
df["Phone_Number"]=df["Phone_Number"].apply(lambda x:str(x))
#applying lambda on the phone and adding '-' on the phone number
df["Phone_Number"]=df["Phone_Number"].apply(lambda x:str(x[0:3]+'-'+x[3:6]+'-'+x[6:10]))
df
```

```python
df[["Street_Address", "City", "Pincode"]] = df["Address"].str.split(',', n=2, expand=True)
df
```

```python
#replacing the 'Yes' and 'No' with 'Y' and 'N'
df["Paying Customer"]=df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"]=df["Paying Customer"].str.replace('No','N')
df
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('No','N')
#removing the  empty spaces
df=df.fillna('')
df
```

```python
#finding the data which can be contacted
for x in df.index:
    if df.loc[x,"Do_Not_Contact"]=="Y":
        df.drop(x,inplace=True)

df
```

```python
df=df.reset_index(drop=True)
df
```
